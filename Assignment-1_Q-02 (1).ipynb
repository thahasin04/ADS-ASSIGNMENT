```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable
```

 1.Create a graph from the karate.gml file and display basic information

```python
karate_graph = nx.read_gml("C:\\Users\\Srinivas\\Downloads\\karate.gml")
print("Basic information about the network:")
print("Number of nodes:", karate_graph.number_of_nodes())
print("Number of edges:", karate_graph.number_of_edges())
print("Graph density:", nx.density(karate_graph))
print("Is connected:", nx.is_connected(karate_graph))
print("Diameter:", nx.diameter(karate_graph))
print("Average shortest path length:", nx.average_shortest_path_length(karate_graph))

# Draw the graph
plt.figure(figsize=(8, 6))
nx.draw(karate_graph, with_labels=True, node_color='skyblue', node_size=500, edge_color='gray', linewidths=0.5)
plt.title("Karate Club Network")
plt.show()
```

2.Store all the metadata of the actors in the network

```python
metadata = {}
for node in karate_graph.nodes(data=True):
    metadata[node[0]] = node[1]

print("\nMetadata of actors in the network:")
print(metadata)

# Draw the graph with metadata as node labels
plt.figure(figsize=(8, 6))
nx.draw(karate_graph, with_labels=True, labels=metadata, node_color='skyblue', node_size=500, edge_color='gray', linewidths=0.5)
plt.title("Karate Club Network with Metadata")
plt.show()
```

3.Calculate all types of centrality (degree, betweenness, closeness, eigenvector, pagerank centrality).

```python
# Analyze the nodes based on the centrality values.
degree_centrality = nx.degree_centrality(karate_graph)
betweenness_centrality = nx.betweenness_centrality(karate_graph)
closeness_centrality = nx.closeness_centrality(karate_graph)
eigenvector_centrality = nx.eigenvector_centrality(karate_graph)
pagerank_centrality = nx.pagerank(karate_graph)

print("\nCentrality measures:")
print("Degree Centrality:", degree_centrality)
print("Betweenness Centrality:", betweenness_centrality)
print("Closeness Centrality:", closeness_centrality)
print("Eigenvector Centrality:", eigenvector_centrality)
print("Pagerank Centrality:", pagerank_centrality)
```

Plot Degree Centrality

```python
print("\nCentrality measures:")
print("Degree Centrality:", degree_centrality)

pos = nx.spring_layout(karate_graph)
nx.draw_networkx_nodes(karate_graph, pos, node_color='pink')
nx.draw_networkx_edges(karate_graph, pos, width=0.5)
nx.draw_networkx_labels(karate_graph, pos, font_size=8)
nx.draw_networkx_edge_labels(karate_graph, pos, edge_labels={e: round(degree_centrality[e[0]], 2) for e in karate_graph.edges()}, label_pos=0.3)
plt.title('Degree Centrality')
plt.show()
```

Plot Betweenness Centrality

4.Find the possible k-components of the network and the clustering coefficient.

```python
k_components = nx.k_components(karate_graph)
clustering_coefficient = nx.average_clustering(karate_graph)

print("\nPossible k-components of the network:", k_components)
print("Clustering Coefficient:", clustering_coefficient)
```

5.Find out the communities using Girvan-Newman algorithm and Louvain method.


```python
communities_girvan_newman = nx.community.girvan_newman(karate_graph)
communities_louvain = nx.community.kernighan_lin_bisection(karate_graph)

girvan_newman_communities = nx.community.girvan_newman(karate_graph)
louvain_communities = nx.community.greedy_modularity_communities(karate_graph)

print("\nCommunities using Girvan-Newman algorithm:", list(girvan_newman_communities))
print("Communities using Louvain method:", list(louvain_communities))
```
